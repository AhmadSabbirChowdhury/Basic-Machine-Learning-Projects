```python
#Linear Regression
from sklearn.model_selection import train_test_split
```

```python
#convert string date to datetime
import datetime as dt
bangladesh_case['date'] = pd.to_datetime(bangladesh_case['date'])
bangladesh_case.head()
```

```python
#converting datetime to ordinal
bangladesh_case['date'] = bangladesh_case['date'].map(dt.datetime.toordinal)
bangladesh_case.head()
```

```python
#Getting dependent variable and independent variable
x = bangladesh_case['date']
y = bangladesh_case['total_cases']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
#Storing the instance in an object
lr = LinearRegression()
```

```python
import numpy as np

#x_train and y_train into "numpy_array" and reshaping them in a single dimention
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
bangladesh_case.tail()
```

```python
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
#x_test will denote all the "actual values" and y_pred will denoted all the "predicted values"
from sklearn.metrics import mean_squared_error
mean_squared_error(x_test,y_pred)
```

```python
## The model we created is not accurate as the mean_squared_error is really big
```
